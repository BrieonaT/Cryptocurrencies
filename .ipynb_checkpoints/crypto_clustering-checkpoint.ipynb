# Clustering Crypto

```python
# Initial imports
import pandas as pd
import hvplot.pandas
from path import Path
import plotly.express as px
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

### Deliverable 1: Preprocessing the Data for PCA

```python
# Load the crypto_data.csv dataset.
file_path = "Resources/crypto_data.csv"
df = pd.read_csv(file_path,index_col=0)
df.head()
```

```python
# Keep all the cryptocurrencies that are being traded.
df = df.loc[df['IsTrading'] == True]
df.head()
```

```python
# Keep all the cryptocurrencies that have a working algorithm.
df = df.dropna(subset = ["Algorithm"])
df.head()
```

```python
# Remove the "IsTrading" column. 
df = df.drop(columns=['IsTrading'])
df.head()
```

```python
# Remove rows that have at least 1 null value.
df = df.dropna()
df.head()
```

```python
# Keep the rows where coins are mined.
df = df.loc[df['TotalCoinsMined'] > 0]
df.head()
```

```python
df.count()
```

```python
# Create a new DataFrame that holds only the cryptocurrencies names.
crypto_df = pd.DataFrame(df["CoinName"])
print(crypto_df.shape)
crypto_df.head()
```

```python
# Drop the 'CoinName' column since it's not going to be used on the clustering algorithm.
df = df.drop(columns=['CoinName'])
df.head()
```

```python
# Use get_dummies() to create variables for text features.
X = pd.get_dummies(df, columns=["Algorithm","ProofType"])

X
```

```python
# Standardize the data with StandardScaler().
X = StandardScaler().fit_transform(X)
X
```

```python
# Using PCA to reduce dimension to three principal components.
# Initialize PCA model
pca = PCA(n_components=3)

X_pca = pca.fit_transform(X)
```

```python
# Create a DataFrame with the three principal components.
pca_df = pd.DataFrame(
                data=X_pca, columns=['PC1', 'PC2', 'PC3'], index = df.index)
pca_df.head()
```

### Deliverable 3: Clustering Crytocurrencies Using K-Means

#### Finding the Best Value for `k` Using the Elbow Curve

```python
# Create an elbow curve to find the best value for K.

inertia = []
k = list(range(1, 11))

#Looking for best K
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(pca_df)
    inertia.append(km.inertia_)

# Define a DataFrame to plot the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k", y="inertia", title="Elbow Curve", xticks=k)
```

Running K-Means with `k=4`

```python
# Initialize the K-Means model.
model = KMeans(n_clusters=4, random_state=0)

# Fit the model
model.fit(pca_df)

# Predict clusters
predictions = model.predict(pca_df)
predictions
```

```python
# Create a new DataFrame including predicted clusters and cryptocurrencies features.
# Concatentate the crypto_df and pcs_df DataFrames on the same columns.
clustered_df = pd.concat([df, pca_df], axis = 1, join = "inner")

#  Add a new column, "CoinName" to the clustered_df DataFrame that holds the names of the cryptocurrencies. 
clustered_df['CoinName'] = crypto_df.CoinName

#  Add a new column, "Class" to the clustered_df DataFrame that holds the predictions.
clustered_df['Class'] = model.labels_

# Print the shape of the clustered_df
print(clustered_df.shape)
clustered_df.head(10)
```

### Deliverable 4: Visualizing Cryptocurrencies Results

#### 3D-Scatter with Clusters

```python
# Creating a 3D-Scatter with the PCA data and the clusters
fig = px.scatter_3d(clustered_df, x="PC1", y="PC2", z="PC3", 
                    color="class", symbol="class", 
                    hover_name='CoinName', hover_data ='Algorithm',
                    width=800)
fig.update_layout(legend=dict(x=0,y=1))

fig.show()
```
